```python
# Kütüphaneleri içe aktarın
import pandas as pd
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
```

```python
# Örnek veri seti oluşturma (semptomlar ve hastalık kategorileri)
data = {
    "symptoms": [
        "Baş ağrısı ve ateş",
        "Mide bulantısı ve karın ağrısı",
        "Boğaz ağrısı ve öksürük",
        "Yorgunluk ve kas ağrıları",
        "Nefes darlığı ve göğüs ağrısı"
    ],
    "label": [0, 1, 2, 3, 4]  # 0: Migren, 1: Gıda Zehirlenmesi, vb.
}

df = pd.DataFrame(data)
print("Veri seti:\n", df)
```

```python
# Veriyi eğitim ve test kümelerine ayırma
X_train, X_test, y_train, y_test = train_test_split(
    df["symptoms"], df["label"], test_size=0.2, random_state=42
)
print("Eğitim veri kümesi:\n", X_train)
```

```python
# Tokenizer ve model yükleme
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=5)

# Veriyi token haline getirme
train_encodings = tokenizer(list(X_train), truncation=True, padding=True, max_length=128, return_tensors="pt")
test_encodings = tokenizer(list(X_test), truncation=True, padding=True, max_length=128, return_tensors="pt")
```

```python
# Eğitim parametreleri ve Trainer ayarı
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_encodings,
    eval_dataset=test_encodings,
)
```

```python
# Modeli eğitme
trainer.train()
```

```python
# Modeli test etme
from transformers import AutoTokenizer

model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)  # Define tokenizer here

sample_inputs = ["Başım ağrıyor ve midem bulanıyor", "Boğazım ağrıyor ve ateşim var"]
inputs = tokenizer(sample_inputs, truncation=True, padding=True, max_length=128, return_tensors="pt")
outputs = model(**inputs)
predictions = outputs.logits.argmax(dim=-1)
print("Tahminler:", predictions)
```

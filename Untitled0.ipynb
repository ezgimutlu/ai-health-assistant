```python
# KÃ¼tÃ¼phaneleri iÃ§e aktarÄ±n
import pandas as pd
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
```

```python
# Ã–rnek veri seti oluÅŸturma (semptomlar ve hastalÄ±k kategorileri)
data = {
    "symptoms": [
        "BaÅŸ aÄŸrÄ±sÄ± ve ateÅŸ",
        "Mide bulantÄ±sÄ± ve karÄ±n aÄŸrÄ±sÄ±",
        "BoÄŸaz aÄŸrÄ±sÄ± ve Ã¶ksÃ¼rÃ¼k",
        "Yorgunluk ve kas aÄŸrÄ±larÄ±",
        "Nefes darlÄ±ÄŸÄ± ve gÃ¶ÄŸÃ¼s aÄŸrÄ±sÄ±"
    ],
    "label": [0, 1, 2, 3, 4]  # 0: Migren, 1: GÄ±da Zehirlenmesi, vb.
}

df = pd.DataFrame(data)
print("Veri seti:\n", df)
```

```python
# Veriyi eÄŸitim ve test kÃ¼melerine ayÄ±rma
X_train, X_test, y_train, y_test = train_test_split(
    df["symptoms"], df["label"], test_size=0.2, random_state=42
)
print("EÄŸitim veri kÃ¼mesi:\n", X_train)
```

```python
# Tokenizer ve model yÃ¼kleme
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=5)

# Veriyi token haline getirme
train_encodings = tokenizer(list(X_train), truncation=True, padding=True, max_length=128, return_tensors="pt")
test_encodings = tokenizer(list(X_test), truncation=True, padding=True, max_length=128, return_tensors="pt")
```

```python
# EÄŸitim parametreleri ve Trainer ayarÄ±
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_encodings,
    eval_dataset=test_encodings,
)
```

```python
# Modeli eÄŸitme
trainer.train()
```

```python
# Modeli test etme
from transformers import AutoTokenizer

model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)  # Define tokenizer here

sample_inputs = ["BaÅŸÄ±m aÄŸrÄ±yor ve midem bulanÄ±yor", "BoÄŸazÄ±m aÄŸrÄ±yor ve ateÅŸim var"]
inputs = tokenizer(sample_inputs, truncation=True, padding=True, max_length=128, return_tensors="pt")
outputs = model(**inputs)
predictions = outputs.logits.argmax(dim=-1)
print("Tahminler:", predictions)
```
